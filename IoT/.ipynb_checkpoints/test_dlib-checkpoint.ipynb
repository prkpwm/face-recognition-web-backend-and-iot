```python
import dlib
import cv2
import os
import numpy as np
```

```python
img = cv2.imread('face_takephoto.bmp')
print(img.shape)
```

```python
detector = dlib.get_frontal_face_detector()
dept = detector(img,0)
print('dept: ',dept[0])
print('len: ',len(dept))
```

```python
predictor_path = os.path.join(os.getcwd(),'model','face_landmarks68.dat')
face_rec_path = os.path.join(os.getcwd(),'model','face_recognition.dat')
```

```python
sp = dlib.shape_predictor(predictor_path) # หาองค์ประกอบของใบหน้า
shape = sp(img,dept[0])
```

```python
print(shape)
```

```python
facerec = dlib.face_recognition_model_v1(face_rec_path)
```

```python
face_descriptor = [i for i in facerec.compute_face_descriptor(img, shape)] # หา 128 จุด
```

```python
len(face_descriptor)
```
